## P.23 Loading the data

Before we do the analysis, the first thing we should do is to load the data from "www.grouplens.org/node/73" and scroll down to find "MovieLens 1M Dataset." Please download "ml-1m.zip." 

"users.dat, movies.dat, ratings.dat" should be moved to your typical path when using python. For Anaconda users, you may move these files to Anaconda file.

```python
import pandas as pd
# Before you write 6 lines, you should take note of changes I made different from the text book.
# First, you should write "engine = 'python'," not to get an error message.
# Second, there are misspelled line. 
# In [13] : (text book) pd.read_table('users.dat') -> (mine) pd.read_table('ratings.dat')
# In [14] : (text book) pd.read_table('users.dat') -> (mine) pd.read_table('movies.dat')
```

```python
unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table('users.dat', sep = '::', header = None, names = unames, engine = 'python')
```

```python
rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('ratings.dat', sep = '::', header = None, names = rnames, engine = 'python')
```

```python
mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('movies.dat', sep = '::', header = None, names = mnames, engine = 'python')
```

```python
# Merging ratings and users, then with movies
data = pd.merge(pd.merge(ratings, users), movies)
data[:10]
```

## P.25

```python
# .ix[] is now changed into .loc[] in python 3.
data.loc[0] # indexing first row
```

```python
# We want to group people by their gender and get the average rating of each movie.
# Take note that python 3 has some change in syntax: row -> index, cols -> columns
mean_ratings = data.pivot_table('rating', index = 'title', columns = 'gender', aggfunc = 'mean')
mean_ratings[:5]
```

```python
# Grouping and counting 
ratings_by_title = data.groupby('title').size()
ratings_by_title[:10]
```

For you to get the grasp of what size() is doing, let me explain some more.
Size() counts how many ratings there are for each of the movie.

```python
# str.contains() search if each of the item in data['title'] has the string you put inside the paranthesis
# and then return boolean (True or False)
data1 = data['title'].str.contains('10 Things I Hate About You')
count = 0 
for i in data1:
    if i == True:  # if it's true, add one to the "count"
        count += 1
    else:
        continue
count
```

Now we find out that those numbers in "ratings_by_title" is frequencies of the ratings for movies.

## P.26

```python
# Looking for movies that got over 250 ratings
active_titles = ratings_by_title.index[ratings_by_title >= 250]

active_titles
```

```python
# loading ratings grouped by gender using index, "active_titles"
mean_ratings = mean_ratings.loc[active_titles]
```

```python
mean_ratings
```

```python
# using .sort_values(), it is possible to sort the movie ratings. 
# .sort_index cannot be used in later version of python.
# But we only want female ratings, which is why we put "by = 'F'"
# For convenience, we need sorting it in descending order, which led us to put "ascending = False" option.
top_female_ratings = mean_ratings.sort_values(by = 'F', ascending = False)
```
